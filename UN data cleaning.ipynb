```python
import pandas as pd

location='https://github.com/HemaVe/PUBPOL542_ComputationAnalytics/raw/main/'
UN_Data=location+ "UNData_ForestProducts_ImportExport.csv"
```

```python
UN_Excel=pd.read_csv(UN_Data)
```

```python
UN_Import = UN_Excel[UN_Excel.Element=='Import Value'] #Filtering the import value
```

```python
UN_Import
```

```python
UN_Import_2019 = UN_Import[UN_Import.Year == 2019]  #Filtering the year 2019
```

```python
 #Filtering out the areas like world and Africa using the value in value footnotes. Im refers to the countries/islands
query1='`Value Footnotes` == "Im"'
UN_Import_2019_Im = UN_Import_2019.query(query1)
```

```python
#Selecting only the country or Area and Value columns
UN_ForestImport_CleanData_new = UN_Import_2019_Im[["Country or Area", "Value"]]
UN_ForestImport_CleanData_new
```

```python
#Cleaning the column names
UN_ForestImport_CleanDatas = UN_ForestImport_CleanData_new.rename(columns={'Country or Area':'Country','Value': 'FPI'})
```

```python
UN_ForestImport_CleanDatas.info()
```

```python
UN_ForestImport_CleanDatas.to_csv('Hemadata2.csv', index = False)
```
